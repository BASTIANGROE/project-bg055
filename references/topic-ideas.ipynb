```python
import requests
from bs4 import BeautifulSoup

name = []
for i in range(5):
    # Send a GET request to the page
    response = requests.get(f"https://battlefieldtracker.com/bf1/leaderboards/stats/all/Kills?page={i+1}")
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, "html.parser")
    
        # Locate the table containing the leaderboard
        leaderboard_table = soup.find("table")  # Adjust selector based on page structure
    
        # Check if the table is found
        if leaderboard_table:
            rows = leaderboard_table.find_all("tr")  # Find all rows in the table
    
            # Loop through each row to extract player data
            for row in rows[1:]:  # Skip the header row
                cols = row.find_all("td")  # Find all columns in the row
                data = [col.text.strip() for col in cols]  # Extract and clean text from each column
                name.append(data[1])  # Print or save the data for each player
        else:
            print("Leaderboard table not found.")
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
